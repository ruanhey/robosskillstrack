```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
urls_hs = ["https://conteudo.movidesk.com/categoria/sucesso-do-cliente/",
           "https://conteudo.movidesk.com/categoria/sucesso-do-cliente/page/2/",
            "https://conteudo.movidesk.com/categoria/sucesso-do-cliente/page/3/"]
```

```python
df = pd.DataFrame(columns = ["url_hs", "nome_artigo", "url_artigo", "duracao"])

for url_hs in urls_hs:
    r = requests.get(url_hs)
    soup = BeautifulSoup(r.content, "html.parser")
    cards = soup.find_all("article", class_ = "posts__item")
    
    for card in cards:
        url_artigo = card.find("h2", class_="featured__title").a["href"]
        nome_artigo = card.find("h2", class_ = "featured__title").find("a").text
        duracao = card.find("div", class_= "featured__category").text.split("| ")[1].replace("\t","")
        
    
        df = df.append(pd.DataFrame([[url_hs, nome_artigo, url_artigo, duracao]],columns = df.columns))
        
df = df.reset_index(drop = True)
```

```python
df_mdesk = pd.DataFrame(columns = ["url_artigo", "img_name", "description", "autor", "data"])

for url_artigo, nome_artigo, index_artigo in zip(df["url_artigo"],df["nome_artigo"],df.index):
    r = requests.get(url_artigo)
    soup = BeautifulSoup(r.content, "html.parser")
    description = soup.find("div", class_ = "ez-toc-v2_0_11 counter-hierarchy counter-decimal").text.replace("\n","")
    autor = soup.find("div", class_ = "name").find("span").text
    data = soup.find("time").text
    img = soup.find("figure", class_ = "post__image").find('img')["src"]
    
    
    path = 'G:/Meu Drive/BRG - SkillTrack/Design/Imagens das ofertas/mdesk'
    img_name = path + "MDesk_" + str(index_artigo) + '.jpg'
    img_data = requests.get(img).content
    with open(img_name, 'wb') as handler:
        handler.write(img_data)
        
    print(url_artigo, description)
        
    df_mdesk = df_mdesk.append(pd.DataFrame([[url_artigo, img_name, description, autor, data]],columns = df_mdesk.columns))
    
```
