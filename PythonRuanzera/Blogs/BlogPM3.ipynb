```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
urls_pm3 = ["https://www.cursospm3.com.br/blog",
            "https://www.cursospm3.com.br/blog?page=2",
            "https://www.cursospm3.com.br/blog?page=3",
            "https://www.cursospm3.com.br/blog?page=4",
            "https://www.cursospm3.com.br/blog?page=5",
            "https://www.cursospm3.com.br/blog?page=6",
            "https://www.cursospm3.com.br/blog?page=7",
            "https://www.cursospm3.com.br/blog?page=8",
            "https://www.cursospm3.com.br/blog?page=9",
            "https://www.cursospm3.com.br/blog?page=10"]
```

```python
df = pd.DataFrame(columns = ["url_pm3", "nome_artigo", "url_artigo"])

for url_pm3 in urls_pm3:
    r = requests.get(url_pm3)
    soup = BeautifulSoup(r.content, "html.parser")
    cards = soup.find_all("div", class_ = "panel blog fluid-video")
    
    for card in cards:
        url_artigo = card.a["href"]
        nome_artigo = card.find("h2", class_ = "blog__title").text
    
        df = df.append(pd.DataFrame([[url_pm3, nome_artigo, "https://www.cursospm3.com.br"+ url_artigo]],columns = df.columns))
        
df = df.reset_index(drop = True)
```

```python
df_pm3 = pd.DataFrame(columns = ["url_artigo", "img_name", "description", "autor", "data"])

for url_artigo, nome_artigo, index_artigo in zip(df["url_artigo"],df["nome_artigo"],df.index):
    r = requests.get(url_artigo)
    soup = BeautifulSoup(r.content, "html.parser")
    post = soup.find("div", class_ = "panel__body").find_all(recursive = False)
    description_start = soup.find("p")
    description_end = soup.find("h4")
    img = soup.find("div", class_ = "panel__body").find('img')["src"]
    data = soup.find("span", class_ = "blog__date").text
    try:
        autor = soup.find("p", class_= "hc hd as he b hf hg hh hi hj hk hl hm hn ho hp hq hr hs ht hu di ff").find("a").text
    except:
        autor = "-"
    try : 
        #post.index(description_end)
        description = "\n".join([tag.text for tag in post[post.index(description_start) : post.index(description_end)]]).replace("\xa0", " ")
    except :
        description = ""
    
    #description = "\n".join([tag.text for tag in post[post.index(description_start) : post.index(description_end)]]).replace("\xa0", " ")
    
    path = 'G:/Meu Drive/BRG - SkillTrack/Design/Imagens das ofertas/'
    img_name = path + "PM3_" + str(index_artigo) + '.jpg'
    img_data = requests.get(img).content
    with open(img_name, 'wb') as handler:
        handler.write(img_data)
        
    df_pm3 = df_pm3.append(pd.DataFrame([[url_artigo, img_name, description, autor, data]],columns = df_pm3.columns))
    
```

```python
df_pm3.merge(df, how = "left", on = "url_artigo").to_excel("Artigos_PM3.xlsx")
```
