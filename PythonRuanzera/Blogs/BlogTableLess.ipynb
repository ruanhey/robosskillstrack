```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
df = pd.DataFrame(columns = ["url_tless", "nome_artigo", "url_artigo", "autor", "descricao"])

r = requests.get("https://tableless.com.br/todos-os-posts/")
time.sleep(15)
soup = BeautifulSoup(r.content, "html.parser")
cards = soup.find_all("div", class_ = "tb-list-item")
url_tless = "https://tableless.com.br/todos-os-posts/"
for card in cards:
    try:
        url_artigo = card.find("a")["href"]
        nome_artigo = card.find("h2").text
        autor = card.find("span", class_ = "tb-author").text.split("por ")[1]
        descricao = card.find("p").text.replace("  ", "").replace("\n", "")
    except :
        autor = ""
        descricao = ""


    df = df.append(pd.DataFrame([[url_tless, nome_artigo, "https://tableless.com.br"+ url_artigo, autor, descricao]],columns = df.columns))

df = df.reset_index(drop = True)
```
