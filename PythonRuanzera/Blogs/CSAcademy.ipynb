```python
df = pd.DataFrame(columns = ["url_artigo", "img_name"])
driver = webdriver.Chrome('chromedriver-Copy1.exe')
driver.get('https://www.csacademy.com.br/blog')
for j in range(0,3):
    driver.find_element_by_class_name("morePosts").click()
    time.sleep(5)

    
cards = driver.find_elements_by_class_name("postArticle")    
for index_card, card in enumerate(cards): 
    url_artigo = card.find_element_by_css_selector('a').get_attribute('href')
    img = driver.find_element_by_class_name("blogImg").find_element_by_css_selector("img").get_attribute('src')
    img_name = "haha"
    path = 'G:/Meu Drive/BRG - SkillTrack/Design/Imagens das ofertas/csa'
    img_name = path + "csa_" + str(index_card) + '.jpg'
    img_data = requests.get(img).content
    with open(img_name, 'wb') as handler:
        handler.write(img_data)

    df = df.append(pd.DataFrame([[url_artigo, img_name]],columns = df.columns))
    
df = df.reset_index(drop = True)
```

```python
df_csa = pd.DataFrame(columns = ["url_artigo", "description", "autor", "url_blog", "nome_artigo", "data"])
driver = webdriver.Chrome('chromedriver-Copy1.exe')
url_blog = "https://www.csacademy.com.br/blog"

for url_artigo, index_artigo in zip (df["url_artigo"], df.index):
    driver.get(url_artigo)
    time.sleep(5)
    driver.execute_script("window.scrollBy(0, document.body.scrollHeight);")
    try:
        data = driver.find_element_by_class_name("u_1131577706").text.split("• ")[1]
        autor = driver.find_element_by_class_name("u_1131577706").text.split("• ")[0]
        post = driver.find_element_by_class_name("u_1471232539").find_element_by_class_name("dmRespCol")
        description_end = driver.find_element_by_css_selector("h2")
        nome_artigo = driver.find_element_by_css_selector("h3").text
        
        description = post.text[0:post.text.find(description_end.text)]

    except :
        description = ""
       
    print(url_artigo, description)   
    df_csa = df_csa.append(pd.DataFrame([[url_artigo, description, autor, url_blog, nome_artigo, data]],columns = df_csa.columns))
```

```python
df_csa.merge(df, how = "left", on = "url_artigo").to_excel("Artigos_CSAcademy.xlsx")
```
