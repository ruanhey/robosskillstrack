```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
urls_hs = ["https://br.hubspot.com/blog/sales/page/1",
            "https://br.hubspot.com/blog/sales/page/2"]
```

```python
df = pd.DataFrame(columns = ["url_hs", "nome_artigo", "url_artigo", "duracao", "skill"])

for url_hs in urls_hs:
    r = requests.get(url_hs)
    soup = BeautifulSoup(r.content, "html.parser")
    cards = soup.find_all("div", class_ = "blog-card__content")
    
    for card in cards:
        url_artigo = card.a["href"]
        nome_artigo = card.find("h3", class_ = "blog-card__content-title").text.replace("\n\n","")
        duracao = card.find("a", class_= "blog-card__read-time").text.split("| ")[1]
        
        try:
            skill = card.find("p", class_ = "blog-card__content-topic").text.replace("\n","")
        except:
            skill = ""
    
        df = df.append(pd.DataFrame([[url_hs, nome_artigo, url_artigo, duracao, skill]],columns = df.columns))
        
df = df.reset_index(drop = True)
```
