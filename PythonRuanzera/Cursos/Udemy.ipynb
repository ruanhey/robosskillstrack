```python
df = pd.DataFrame(columns = ["url_cursos"])
driver = webdriver.Chrome('chromedriver.exe')
for j,curso in enumerate(cursos):
    driver.get('https://www.udemy.com/')
    search_input = driver.find_element_by_class_name('js-header-search-field')
    # let google find any linkedin user with keyword "python developer" and "San Francisco"
    search_input.send_keys( curso )
    search_input.send_keys(Keys.RETURN)
    time.sleep(3)
    url_cursos = driver.find_element_by_class_name("course-list--container--3zXPS").find_element_by_class_name("popover--popover-hover--14ngr").find_element_by_css_selector('a').get_attribute('href')
    print(url_cursos)
    df = df.append(pd.DataFrame([[url_cursos]],columns = df.columns))
df = df.reset_index(drop = True)
```

```python
df_udemy = pd.DataFrame(columns = ["formato","nome", "instituicao", "site_plataforma", "conteudo","professor","descricao","nivel", "rating", "skills", "data", "duracao", "preco","periodo", "idioma", "modalidade", "img_name"])
driver = webdriver.Chrome('chromedriver.exe')
for url_curso, index_curso in zip (df["url_cursos"], df.index):
    driver.get(url_curso)
    time.sleep(3)
    div_conteudos = driver.find_element_by_class_name("curriculum--content--PkBXH").find_elements_by_class_name("section--section-title--8blTh")
    conteudo = "\n".join([conteudo.text for conteudo in div_conteudos])
    formato = "Curso Livre Online"
    nome = driver.find_element_by_class_name("clp-lead__title").text
    instituicao = "Udemy"
    site_plataforma = "https://www.udemy.com/"
    professor = driver.find_element_by_class_name("instructor--instructor__title--34ItB").find_element_by_css_selector("a").text
    descricao = driver.find_element_by_class_name("show-more--content--isg5c").find_element_by_css_selector("div").text
    nivel = "-"
    rating = driver.find_element_by_class_name("review-summary-widget--average-number--2uYPM").text +"/5"
    skills =""
    data ="-"
    duracao = driver.find_element_by_xpath('//span[text()="Duração total: "]').text.split(": ")[1]
    spans = [s.text for s in driver.find_elements_by_tag_name("span")]
    regex = re.compile(r'^R\$(.*)')
    preco = [i for i in spans if regex.match(i)][0]
    periodo = "-"
    idioma = driver.find_element_by_class_name("clp-lead__element-meta").find_element_by_class_name("clp-lead__locale").text
    modalidade = "Online"
    img = driver.find_element_by_class_name("intro-asset--img-aspect--1UbeZ").find_element_by_tag_name("img").get_attribute("src")
    path = 'G:/Meu Drive/BRG - SkillTrack/Design/Imagens das ofertas/Udemy'
    img_name = path + "Udemy_" + str(index_curso) + '.jpg'
    img_data = requests.get(img).content
    with open(img_name, 'wb') as handler:
        handler.write(img_data)
    print(preco)
    df_udemy = df_udemy.append(pd.DataFrame([[formato, nome, instituicao, site_plataforma, conteudo, professor, descricao, nivel, rating, skills, data, duracao, preco, periodo, idioma, modalidade, img_name]],columns = df_udemy.columns))
```
