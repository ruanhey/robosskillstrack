```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
urls_spotify = [ "https://open.spotify.com/show/7uRmxP1l3m9kwrsTYxnIMA"]
```

```python
df = pd.DataFrame(columns = ["nome_podcast", "autor", "nome_episodio","url_episodio", "data", "duracao"])

for url_spotify in urls_spotify:
    r = requests.get(url_spotify)
    time.sleep(20)
    soup = BeautifulSoup(r.content, "html.parser")
    episodios = soup.find_all("li", class_= "tracklist-row")
    nome_podcast = soup.find("div", class_ = "media-bd").find("h1").text
    autor = soup.find("div", class_ ="media-bd").find("h2").text
    
    for episodio in episodios:
        nome_episodio = episodio.find("span", class_ = "track-name").text
        url_episodio = "https://open.spotify.com" + episodio.a["href"]
        data = episodio.find("span", class_ = "artists-albums").text
        duracao = episodio.find("span", class_ = "total-duration").text.split(":")
        if (len(duracao)==2):
            duracao = duracao[-2]+"min"
        if (len(duracao)==3):
            duracao = duracao[-3]+"h" + duracao[-2]+"min"
        
        df = df.append(pd.DataFrame([[nome_podcast, autor, nome_episodio, url_episodio, data, duracao]],columns = df.columns))
    
```

```python
df_spotify = pd.DataFrame(columns = ["url_episodio", "descricao", "img_name"])

for url_episodio in df["url_episodio"]:
    r = requests.get(url_episodio)
    time.sleep(5)
    soup = BeautifulSoup(r.content, "html.parser")
    try : 
        descricao = soup.find("p", class_= "entity-long-description").text
    except :
        descricao = ""
    if (soup.find_all('img')[1]["src"][:5]=="https"):
        img = soup.find_all('img')[1]["src"]
    else: 
        img = "https:" + soup.find_all('img')[1]["src"]
    path = 'G:/Meu Drive/BRG - SkillTrack/Design/Imagens das ofertas/pg'
    img_name = path + "Podcast" + url_episodio.split("/")[-1][0:5] + '.jpg'
    img_data = requests.get(img).content
    with open(img_name, 'wb') as handler:
        handler.write(img_data)
        
    df_spotify = df_spotify.append(pd.DataFrame([[url_episodio, descricao, img_name]],columns = df_spotify.columns))
```

```python
df_spotify.merge(df, how = "left", on = "url_episodio").to_excel("Podcasts.xlsx")
```
